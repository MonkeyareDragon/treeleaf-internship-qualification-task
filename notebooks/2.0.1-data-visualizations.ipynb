```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Path to save
file_path = '../data/processed/final_bank_loan_data.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
# Set the style for the plots
sns.set(style="whitegrid")

# Distribution of Age, Income, and Family Size
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Age distribution
sns.histplot(df['Age'], bins=20, kde=True, ax=axes[0])
axes[0].set_xlim(0, 100)
axes[0].set_title('Age Distribution')

# Income distribution
sns.histplot(df['Income'], bins=20, kde=True, ax=axes[1])
axes[1].set_title('Income Distribution')

# Family size distribution
sns.histplot(df['Family'], bins=5, kde=True, ax=axes[2])
axes[2].set_title('Family Size Distribution')

plt.tight_layout()
plt.show()
```

```python
# Relationship between Income and Credit Card Average (CCAvg)
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Income', y='CCAvg', hue='Personal Loan', data=df, palette='viridis')
plt.title('Income vs Credit Card Average')
plt.show()
```

```python
# Comparison of Income and Mortgage by Personal Loan status
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Income by Personal Loan
sns.boxplot(x='Personal Loan', y='Income', data=df, ax=axes[0])
axes[0].set_title('Income by Personal Loan')

# Mortgage by Personal Loan
sns.boxplot(x='Personal Loan', y='Mortgage', data=df, ax=axes[1])
axes[1].set_title('Mortgage by Personal Loan')

plt.tight_layout()
plt.show()
```

```python
# Education Level Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x='Education', data=df)
plt.title('Education Level Distribution')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 10))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
